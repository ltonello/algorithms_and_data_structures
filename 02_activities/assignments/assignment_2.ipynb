```python
# Import List from typing for type hinting
from typing import List
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, val = 0, left = None, right = None):
        self.val = val
        self.left = left
        self.right = right

def bt_path(root: TreeNode) -> List[List[int]]:
    # If the root is None, return an empty list
    if not root:
        return []
    
    # Initialize the result list to store all paths
    paths = []
    
    # Helper function to perform depth-first search
    def dfs(node, current_path):
        # If the node is None, return
        if not node:
            return
        
        # Add current node's value to the path
        current_path.append(node.val)
        
        # If this is a leaf node (no children), add the path to results
        if not node.left and not node.right:
            paths.append(current_path.copy())
        
        # Recursively explore left and right subtrees
        if node.left:
            dfs(node.left, current_path)
        if node.right:
            dfs(node.right, current_path)
        
        # Backtrack by removing the last node
        current_path.pop()
    
    # Start the depth-first search from the root
    dfs(root, [])
    
    return paths
```


-   Explain why their solution works in your own words.


```python
# Their solution works because each time a path is found, it is stored in a list and the paths are kept track of there. They are also using a depth-first helper function that
# ensures that all nodes are visited and all paths are accounted for.
```


-   Explain the problem’s time and space complexity in your own words.


```python
# The time complexity for this problem is O(n) where n is the number of nodes. Since each node is visited once and the actions done on each node is constant (O(1)). 
# The time complexity is O(n).

# The space complexity for this problem is O(n*h) as the maximum space for holding nodes is all the nodes in the tree (n) and the maximum space for holding a path is 
# the height of the tree (h).
```
